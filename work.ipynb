#### Box & Whisker Plot

Useful for visualizing the spread & skew of data.

The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.

So, half of the data exists within the box.

The dotted-line "whiskers" indicate the range of the data - except for outliers, which are plotted outside the whiskers. Outliers are 1.5X or more the interquartile range.

This example below creates uniformly distributed random numbers between -40 and 60, plus a few outliers above 100 and below -100:

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```

# Covariance and correlation

##### Covariance
It measures how two variables vary in tandem from their means.

Interpreting covariance: we know a small covariance, close to 0, means there isn't much correlation between the two variables. And large covariance - that is far from o (could be negativ for inverse relationship) - mean there is a correlation:

- correlation of -1 means a perfect inverse correlation
- correlation of 0 means no correlation
- correlation of 1 means perfect correlation

Remember: correlation does not imply causation

You can think conceptually at the covariance as sort of taking these multidimensional vectors of variances from the mean for each attribute and computing the angle between theme as a measure of the covariance. 

When we calculate the covariance, basically we treat each variable as a vector of deviations from the mean, and compute the "dot product" of both vectors. Geometrically, this can be thought of as the angle between the two vectors in a high dimensional space, but you can just think of it as a misure of similiraty between the two variables.

First, let's just make page speed and purchase amount totally random and indepndent of each other; a very smal covariance will result as there is no real correlation:

```python
import numpy as np
```

```python
# compute the deviation from the mean for each datapoint
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]
```

```python
def covariance(x,y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1)
```

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)
```

```python
import matplotlib.pyplot as plt
plt.scatter(pageSpeeds, purchaseAmount)
```

```python
covariance(pageSpeeds, purchaseAmount)
```

Now we will make our fabricated purchase amounts an actual function of page speed, making a very real correlation. The negative value indicated an inverse relationship: pages that render in less time result in more money spent:

```python
purchaseAmount = np.random.normal(30, 10, 1000) / pageSpeeds
```

```python
plt.scatter(pageSpeeds, purchaseAmount)
```

```python
covariance(pageSpeeds, purchaseAmount)
```

What does this value mean? Covariance is sensitive to the units used in the variables, which makes it difficult to interpret. Correlation normalizes everything byt their standard deviations, giving you an easier to understand value that range from -1 to 1

```python
# Normalization of the covariance
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy  #In real life you'd check for divide by zero here

correlation(pageSpeeds, purchaseAmount)
```

numpy can do all this for you with numpy.corrcoef. It returns a matrix of the correlation coefficients between every combination of the arrays passed in:

```python
numpy.corrcoef(pageSpeeds, purchaseAmount)
```

We can force a perfect correlation by fabricating a totally linear relationship (again, it's not exactly -1 just due to precision errors, but it's close enough to tell us there's a really good correlation here):

```python
purchaseAmount = 100 - pageSpeeds * 3

plt.scatter(pageSpeeds, purchaseAmount)

correlation (pageSpeeds, purchaseAmount)
```

# Conditional probability

Bayes formula

Below is some code to create some fake data on how much stuff people purchase given their age range.

It generates 100,000 random "people" and randomly assigns them as being in their 20's, 30's, 40's, 50's, 60's, or 70's.

It then assigns a lower probability for young people to buy stuff.

In the end, we have two Python dictionaries:

"totals" contains the total number of people in each age group.
"purchases" contains the total number of things purchased by people in each age group.
The grand total of purchases is in totalPurchases, and we know the total number of people is 100,000.

Let's run it and have a look:

```python
from numpy import random

random.seed(0) # the purpose is to give consistene each time the code is run

#dictionaries
totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}

totalPurchases = 0

for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = 0.4
    #purchaseProbability = float(ageDecade) / 100
    totals[ageDecade]  += 1
    if(random.random() < purchaseProbability): 
        totalPurchases += 1
        purchases[ageDecade] += 1
```

```python
totals
```

```python
purchases # as we see that the data we can identify a trend
```

```python
totalPurchases
```

Let's play with conditional probability.

First let's compute P(E|F), where E is "purchase" and F is "you're in your 30's". The probability of someone in their 30's buying something is just the percentage of how many 30-year-olds bought something:

```python
PEF = float(purchases[30]) / float(totals[30])
print("P(purchase | 30s): " + str(PEF))
```

```python
# PF - the probability of being 30 in the dataset
PF = float(totals[30]) / 100000
print("P(30's): " + str(PF))
```

```python
# The overall probability of buying something, regardless of your age
PE = float(totalPurchases)/100000
print("P(Purchase): " +  str(PE))
```

If E and F were independent, then we would expect P(E | F) to be about the same as P(E). But they're not; P(E) is 0.45, and P(E|F) is 0.3. So, that tells us that E and F are dependent (which we know they are in this example.)

P(E,F) is different from P(E|F). P(E,F) would be the probability of both being in your 30's and buying something, out of the total population - not just the population of people in their 30's:

```python
print("P(30's, Purchase): "+str(float(purchases[30])/100000))
```

Let's also compute the product of P(E) and P(F), P(E)P(F):

```python
print("P(30's)P(Purchase): " + str(PE * PF))
```

Something you may learn in stats is that P(E,F) = P(E)P(F), but this assumes E and F are independent. We've found here that P(E,F) is about 0.05, while P(E)P(F) is about 0.075. So when E and F are dependent - and we have a conditional probability going on - we can't just say that P(E,F) = P(E)P(F).

We can also check that P(E|F) = P(E,F)/P(F), which is the relationship we showed in the slides - and sure enough, it is:

```python
print((purchases[30] / 100000.0) / PF)
```

# Linear Regression


Fit a line to a  data set of observations and use this line to predict unobserved values.

How does it works? Usually using "least squares", it minimizes the squared-error between each point and the line. 
This is the same as maximizing the likelihood of the observed data if you start thinking of the problem in terms of probabilities and probability distribution functions. This is sometimes called "maximum likelihood estimation"

##### More than one way to do it:
Gradient Descent in an alternative method to least squares. Basically iterates to find the line that best follows the contours defined by the data. 
Can make sense when dealing with 3d data.

###### Measuring error with r-squared 
It is how to measure how well the line fits the data. R-squared mesures: the fraction of the total variation in Y that is captured by the model i.e. 1 - sumOfSquaredErros/sumOfSquaredVariationFromMean

Interpretation r-squared:
    - ranges from 0 to 1
    - 0 is bad (none of the variance is captured)
    - 1 os good (all of the variance is captured

```python
pageSpeeds = np.random.normal(3, 1, 1000)
```

```python
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0,0.1,1000)) * 3

plt.scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
# The R-squared (r_value is the correlation coefficient) shows a really good fit:
r_value ** 2
```

Let's use the slope and intercept we got from the regressopm to plot predicted values vs. observed

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept # the line 

fitLine = predict(pageSpeeds)
```

# Bayesian Methods - Naive bayes for Spam classifier

Naive Bayes - assume indipendecies between corelated events. Example: the words typical for a spam email considered as indipendent one from each other

```python
import os 
import io
import numpy
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

def readFiles(path):
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            path = os.path.join(root, filename)

            inBody = False
            lines = []
            f = io.open(path, 'r', encoding='latin1')
            for line in f:
                if inBody:
                    lines.append(line)
                elif line == '\n':
                    inBody = True
            f.close()
            message = '\n'.join(lines)
            yield path, message


def dataFrameFromDirectory(path, classification):
    rows = []
    index = []
    for filename, message in readFiles(path):
        rows.append({'message': message, 'class': classification})
        index.append(filename)

    return DataFrame(rows, index=index)
        
data = DataFrame({'message': [], 'class': []})

data = data.append(dataFrameFromDirectory('C:\\users\\Alex\\Desktop\\Untitled Folder\\MLCourse\\emails\\spam','spam'))
data = data.append(dataFrameFromDirectory('C:\\Users\\Alex\\Desktop\\Untitled Folder\\MLCourse\\emails\\ham','ham'))
                   
```

# K-Means Clustering

Attempts to split data into K groups that are closest to K centroids. 


Hoe it works:
- randomly pick K centroids
- assign each data point to the centroid it's cloeses to
- Recompute the centroids based on the averge position of each centroid's points
- iterate until points stop changing assignement to centroids

At the end, if you want to predicts the cluster for new points, just find the centroid they're closest to

K-means clustering gotachas:
- choosing k: try increasing K values until you stop getting large reductions in squared error (distrances from each point to their centroids)
- avoiding local minima: the random choice od initial centroids can yield different results. Therefore, run it a few times just to make sure your initial results aren't wacky
- labeling the clusters: K-Means does not attempt to assign any meaning to the clusters you find. It's up to you to dig into the data and try to determine that

Let's make some fake data that includes people clustered by income and age, randomly:

```python
from numpy import random, array

#Create fake income/age clusters for N people in k clusters
def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    for i in range (k):
        incomeCentroid = random.uniform(20000.0, 200000.0)
        ageCentroid = random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
    X = array(X)
    return X
```

We'll use k-means to rediscover these clusters in unsupervised learning:

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float

data = createClusteredData(100, 15)

model = KMeans(n_clusters=5)
```

# Ensemble Learning

Random forest - it use multiple models to try and solve the same problem

- bagging: many models are buit bytraining on randomly-drawn subset of the data. Ex. Random Forest
- Boosting: alternative technique where each subsequent model in the ensemble boosts attribute that address data mis-classified by the previous model
- bucket of models: it trains several different models using training data, and picks the one that works best with test data
- Stacking: it runs multiple models at once on the data, and combines thre results together

Advanced ensemble learning (to look smart):
- Bayes Optimal classifier: theoretically the best, but almost impractical

- Bayesian Model Combination: it's about the same as using cross-validation to find the best combiantion of models

# Support Vector Machines

It works well for classifying higher-dimensional data (ie. lots of feature)

Finds higher-dimensional support vectors across which to divide the data

Uses something called the kernel trick to represent data in higher-dimesional spaces to find hyperplanes that might not be apparent  in lower dimensions

Let's create the same fake income/age clustered data that we used for our K-Means clusterings

```python
import numpy as np

#Create fake income/age clusters for N people in k clusters
def createClusteredData(N, k):
    np.random.seed(1234)
    pointsPerCluster = float(N)/k
    X = []
    y = []
    for i in range (k):
        incomeCentroid = np.random.uniform(20000.0, 200000.0)
        ageCentroid = np.random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            X.append([np.random.normal(incomeCentroid, 10000.0), np.random.normal(ageCentroid, 2.0)])
            y.append(i)
    X = np.array(X)
    y = np.array(y)
    return X, y
```

```python
%matplotlib inline
from pylab import *
from sklearn.preprocessing import MinMaxScaler

(X, y) = createClusteredData(100, 5)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y.astype(np.float))
plt.show()
```

```python
# Scalling of the values, this is better for SVM
scalling = MinMaxScaler(feature_range=(-1,1)).fit(X)
X = scalling.transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y.astype(np.float))
plt.show()
```

Now we'll use linear SVC to portion our graph into cluster:

```python
from sklearn import svm, datasets

C = 1 #Regularization
svc = svm.SVC(kernel = 'linear', C=C).fit(X, y)
```

By settiing up a dense mesh of points in the grid and classifying all of them, we can render the regions of each cluster as dinsincs colors

```python
def plotPredictions(clf):
    # Create a dense grid of points to sample 
    xx, yy = np.meshgrid(np.arange(-1, 1, .001),
                     np.arange(-1, 1, .001))
    
    # Convert to Numpy arrays
    npx = xx.ravel()
    npy = yy.ravel()
    
    # Convert to a list of 2D (income, age) points
    # this instruction concatenate the points
    samplePoints = np.c_[npx, npy]
    
    # Generate predicted labels (cluster numbers) for each point
    Z = clf.predict(samplePoints)

    plt.figure(figsize=(8, 6))
    Z = Z.reshape(xx.shape) #Reshape results to match xx dimension
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8) # Draw the contour
    plt.scatter(X[:,0], X[:,1], c=y.astype(np.float)) # Draw the points
    plt.show()
    
plotPredictions(svc)
```
